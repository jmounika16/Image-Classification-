```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization, Activation
from keras.constraints import maxnorm
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras.datasets import cifar10
```

```python
# fix random seed for reproducibility
seed = 21
np.random.seed(seed)

#Dropout percentage 
percentage = .15

# load data
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# normalize inputs from 0-255 to 0.0-1.0
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train = X_train / 255.0
X_test = X_test / 255.0

# one hot encode outputs, creates numerical categorical variables
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]

# Create the model
model = Sequential()

#adds a convolution layer
model.add(Conv2D(32, (6, 6), input_shape=X_train.shape[1:], padding='same'))
#creates the non linear model of the convolution layer
model.add(Activation('relu'))
#drops a percentage of data to ensure results are not overfit
model.add(Dropout(percentage))
#ensure that results are distributed properly before adding next layer
model.add(BatchNormalization())

model.add(Conv2D(64, (6, 6), padding='same'))
model.add(Activation('relu'))
#Pools important features learned from model
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(percentage))

model.add(BatchNormalization())

model.add(Conv2D(64, (6, 6), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(percentage))
model.add(BatchNormalization())

model.add(Conv2D(128, (6, 6), padding='same'))
model.add(Activation('relu'))
model.add(Dropout(percentage))
model.add(BatchNormalization())

#transformed layers into vectors
model.add(Flatten())
model.add(Dropout(percentage))

#creates desenley collected layer, based on number of neurons
model.add(Dense(256, kernel_constraint=maxnorm(3)))
model.add(Activation('relu'))
model.add(Dropout(percentage))
model.add(BatchNormalization())
model.add(Dense(128, kernel_constraint=maxnorm(3)))
model.add(Activation('relu'))
model.add(Dropout(percentage))
model.add(BatchNormalization())
model.add(Dense(num_classes))

#chooses the results based on the highest percentages
model.add(Activation('softmax'))

epochs = 25
optimizer = 'Adam'

model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

print(model.summary())

model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=64)

# Final evaluation of the model

scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
